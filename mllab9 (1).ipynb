```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()

dtree.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

dtree_pred = dtree.predict(X_test)

print(confusion_matrix(y_test, dtree_pred))
print(classification_report(y_test, dtree_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=200)

rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, rfc_pred))
```

^ No warning appeared, but the model performs poorly for class 1.
This is due to class imbalance in the dataset. The model shows high accuracy but performs poorly in predicting class 1 (not fully paid).
This is due to class imbalance, where most of the data belongs to class 0.
As a result, the model predicts the majority class more often and fails to correctly classify minority cases.

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, rfc_pred))
```

The Random Forest achieved higher overall accuracy compared to the Decision Tree.
However, the Decision Tree performed better in predicting the minority class (not fully paid), while the Random Forest struggled significantly with this class due to class imbalance.
